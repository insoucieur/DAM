Before doing anything else, let's load some packages.

The line "%matplotlib inline" enables you to display plots within the iPython notebook

The package numpy contains mathematical operations; matplotlib.pyplot contains plotting options.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

The file 'lonstatistik.txt' contains the average disposable income for men and women, respectively, in 108 different areas of Denmark in 2014 (source: Danmarks Statistik). Let's open it.

```python
data = np.loadtxt('lonstatistik.txt')
```

View the dimensions of the 'data matrix':

```python
data.shape
```

Let's store the incomes of men and women in two different vectors.

```python
men = data[:,0]
women = data[:,1]
```

Let's start out by plotting the men's incomes as a function of the 108 indexed regions.

```python
region_indices = range(108)
plt.plot(region_indices, men)
#plt.xlabel('Region index')
#plt.ylabel('Disposable income in DKK')
#plt.title('Disposable income for 108 different regions')
```

In order to compare with the women's incomes, let's plot both together

```python
plt.plot(region_indices, men, label='Men') # Add the labels to label the graphs
plt.plot(region_indices, women, color='red', label='Women') # Add the labels to label the graphs
plt.xlabel('Region index')
plt.ylabel('Disposable income in DKK')
plt.title('Disposable income for 108 different regions')
plt.legend() # Include the legend marking the two graphs
```

A more suitable plot might have been a *bar plot*

```python
plt.bar(region_indices, men, label='Men')
plt.bar(region_indices, women, color='red', label='Women')
plt.xlabel('Region index')
plt.ylabel('Disposable income in DKK')
plt.title('Bar plot over disposable incomes for women and men')
plt.legend()
```

Next, make a *scatter plot* of the incomes of women versus men

```python
plt.scatter(women, men, label='Regional salaries')
plt.title('Disposable income for men versus women in 108 different areas of Denmark')
plt.xlabel('Women\'s disposable income in DKK')
plt.ylabel('Men\'s disposable income in DKK')
plt.axis([150000,500000, 150000, 500000]) # Make axes equal for a more accurate description
plt.plot([150000, 500000], [150000, 500000], color='red', label='Equal salaries') 
# Adding in a diagonal line illustrating where equal salaries would be found
plt.legend()
```

Next, let's make *histograms* over the income values of men and women. Histograms estimate the distribution over incomes by counting occurrences of different *bins* of incomes

```python
plt.hist(men, 20, label='Men')
plt.hist(women, 20, color='red', label='Women')
plt.xlabel('Disposable income in DKK')
plt.ylabel('Count')
plt.title('Histogram over income levels in women and men')
plt.legend()
```

Finally, let's make box plots for the two populations of incomes. The red line in the box plot shows the median income, the borders of the blue box show the first and third quartiles, and the whiskers represent the range of the data. More about what that means in Lecture 3!

```python
labels = ['Men', 'Women']
plt.boxplot(data, labels=labels)
plt.title('Boxplot over disposable income levels for men and women')
plt.ylabel('Disposable income')
```
